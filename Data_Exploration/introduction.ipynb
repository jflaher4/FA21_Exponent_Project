```python
x = 'OverallQual'
y = 'SalePrice'
plt.scatter(df[x],df[y])
plt.title(f'{x} vs. {y} Scatterplot')
plt.xlabel(x)
plt.ylabel(y)
plt.show()
df[x].describe()
```

```python
df[[x,y]].groupby(x).mean()
```

```python
df[['GrLivArea','TotalBsmtSF']].head(5)
```

```python
df['TotalSF'] = df['GrLivArea'] + df['TotalBsmtSF']
df['TotalSF'].head(5)
```

```python
x = 'TotalSF'
y = 'SalePrice'
coef = np.polyfit(df[x],df[y],1)
poly1d_fn = np.poly1d(coef)
plt.plot(df[x],df[y],'go',df[x],poly1d_fn(df[x]),':y')
plt.title(f'{x} vs. {y} Scatterplot')
plt.xlabel(x)
plt.ylabel(y)
plt.show()
df[x].describe()
```

```python
linregress(df['TotalBsmtSF'],df[y]).rvalue
linregress(df['GrLivArea'],df[y]).rvalue
linregress(df[x],df[y]).rvalue
```
